Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.
<p>In Pandas, a DataFrame is a two-dimensional table-like data structure, where each column can have a different data type. It is similar to a spreadsheet or SQL table and can be used to store and manipulate data. A DataFrame can be thought of as a collection of Pandas Series, where each column in the DataFrame represents a Pandas Series.

</p>
<p>
On the other hand, a Pandas Series is a one-dimensional labeled array capable of holding any data type. It is similar to a Python list or array, but with added functionality and flexibility.
</p>

```python
# Series
age = [12,23,34,43,45,43,32,21,54]
data = pd.Series(age)
print(data)
```

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
        'age': [25, 32, 18, 47, 22],
        'country': ['USA', 'Canada', 'USA', 'USA', 'Canada']}
df = pd.DataFrame(data)
print(df)
```

Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

<p>
In Pandas, both Series and DataFrame are mutable in nature, which means that you can modify their contents after they have been created.

A Series is a one-dimensional labeled array that can hold any data type. You can modify the values in a Series by assigning new values to specific indices.

A DataFrame is a two-dimensional labeled data structure with columns of potentially different types. You can modify the values in a DataFrame by assigning new values to specific rows and columns.

On the other hand, a Panel is a three-dimensional labeled array that can hold any data type. However, Panels are not recommended for most use cases and have been deprecated in the latest version of Pandas (1.2.0). Instead, you can use multi-dimensional arrays or DataFrames to represent higher-dimensional data.
</p>

Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# Creating a Pandas Series for each column
names = pd.Series(['Alice', 'Bob', 'Charlie', 'David', 'Emily'])
ages = pd.Series([25, 32, 18, 47, 22])
countries = pd.Series(['USA', 'Canada', 'USA', 'USA', 'Canada'])

# Combining the Series into a DataFrame
df = pd.DataFrame({'Name': names, 'Age': ages, 'Country': countries})
print(df)

# Output:
#        Name  Age Country
# 0     Alice   25     USA
# 1       Bob   32  Canada
# 2   Charlie   18     USA
# 3     David   47     USA
# 4     Emily   22  Canada
```
